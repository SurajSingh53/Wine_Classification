A **Decision Tree** is a non-parametric supervised learning algorithm, which is utilized for both classification and regression tasks. It has a hierarchical, tree structure, which consists of a root node, branches, internal nodes and leaf nodes.

Here we Will be using **Wine Dataset** to depict the working of the Decision tree model. 

1. Importing all the necessary libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets,tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

2.Loading the dataset

```python
dataset = datasets.load_wine()
x  = dataset.data
y = dataset.target
```

3. Splitting the dataset into train and test set ( 80:20 ration here ).

```python
x_train,x_test,y_train,y_test = train_test_split(x , y ,train_size=0.8, random_state= 42 )
```

4. Scaling the Input data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_test = scaler.fit_transform(x_test)
x_train = scaler.fit_transform(x_train)
```

5. Applying **Decision tree Classifier** model and fitting the training data 

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

6. Predicting the output using x_test as input.

```python
y_pred = model.predict(x_test)
y_pred
```

7. checking the accuracy of the model.

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

8. Plotting Confusion Matrix of the model.

```python
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
cm = confusion_matrix(y_pred,y_test)
sns.heatmap(cm , annot = True ,fmt= 'd', cbar=False).set(xlabel = 'Predicted', ylabel = 'Actual')
```

```python
print(classification_report(y_pred,y_test),"\n")
```

9. Now we are printing the model in Text format.

```python
represent = tree.export_text(model)
print(represent,end='\n')
```

10. Creating and Saving the Graphical reprsentaion of the model.

```python
fig = plt.figure(figsize=(20,20))
treeplot = tree.plot_tree(model , feature_names= dataset.feature_names , class_names= dataset.target_names , filled=True)
```

```python
fig.savefig("Wine_Decisiontree.png")
```
